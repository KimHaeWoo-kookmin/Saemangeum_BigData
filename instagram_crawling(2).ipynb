```python
# ê²€ìƒ‰ì°½ì—ì„œ ì²«ë²ˆì§¸ ê²Œì‹œë¬¼ ì„ íƒ

def select_first(driver) :
    first = driver.find_element(By.CLASS_NAME,"_aagw")
    first.click()
    time.sleep(3)
```

```python
# ê²Œì‹œë¬¼ì— ë“±ì¥í•˜ëŠ” [ë³¸ë¬¸ ë‚´ìš©, í•´ì‹œíƒœê·¸, ì‘ì„±ì¼ì, ì¢‹ì•„ìš”, ìœ„ì¹˜]ë¥¼ Htmlíƒœê·¸ë¥¼ í™œìš©í•˜ì—¬ í¬ë¡¤ë§

def get_content(driver) :
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    #ë³¸ë¬¸ ë‚´ìš©
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''
    #í•´ì‹œíƒœê·¸
    tags = re.findall(r'#[^\s#,\\]+', content)
    
    #ì‘ì„±ì¼ì
    date = soup.select('time._aaqe')[0]['datetime']
    
    #ì¢‹ì•„ìš”
    try:
        like = soup.find('span', {'class':'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj'}).text
    except:
        like = 0
    #ìœ„ì¹˜
    try:
        place = soup.select('div._aaqm')[0].text
    except:
        place = ''
        
    data = [content, date, like, place, tags]
    return data
```

```python
# í•œ ê²Œì‹œë¬¼ì˜ í¬ë¡¤ë§ì´ ëë‚˜ë©´ ë‹¤ìŒ ê²Œì‹œë¬¼ë¡œ ì´ë™

def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR, "div._aaqg._aaqh")
    right.click()
    time.sleep(3)
```
